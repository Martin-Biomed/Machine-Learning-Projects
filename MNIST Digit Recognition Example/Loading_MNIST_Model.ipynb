```
from __future__ import absolute_import, division, print_function

import tensorflow as tf
from tensorflow.keras import Model, layers
import numpy as np
import matplotlib.pyplot as plt

num_features = 784 # data features (img shape: 28*28).
```

```
from google.colab import drive
drive.mount('/content/gdrive')
```

```
# We do not need to compile the model, we only need to compile when we want to train the model more
# Here, we only want to use the model that has already been trained

SAVED_MODEL_LOCATION = "/content/gdrive/MyDrive/assets/MNIST_Basic_Neural_Network"

reloaded_model = tf.keras.models.load_model(SAVED_MODEL_LOCATION, compile=False)
reloaded_model.summary()
```

```
# I used an individual image (arbitrarily chosen, you may choose other ones) to test the prediction of this model
image_9 = plt.imread('/content/gdrive/MyDrive/ML_Datasets/Individual_MNIST_Images/9_28pixel.jpg') # Where I was keeping this individual image

#Converting the image to float32
image_9_array = np.array(image_9, np.float32)

plt.imshow(image_9)
```

```
# Flatten images to 1-D vector of 784 features (28*28).
image_9_array = image_9_array.reshape([-1, num_features])

# Normalize images value from [0, 255] to [0, 1].
image_9_array = image_9_array/255.
```

```
#We now pass the image to the model we trained before
prediction_9 = reloaded_model(image_9_array)

plt.imshow(np.reshape(image_9, [28, 28]), cmap='gray') 
print("Model prediction: %i" % np.argmax(prediction_9.numpy()))  #This will output the prediction from our model
plt.show()
```
