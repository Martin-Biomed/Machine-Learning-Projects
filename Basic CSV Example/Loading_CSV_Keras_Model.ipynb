```
# This notebook shows how to use a previously saved h5 model to input new values.

import pandas as pd
import numpy as np

# Make numpy values easier to read.
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
```

```
abalone_train = pd.read_csv(
    "https://storage.googleapis.com/download.tensorflow.org/data/abalone_train.csv",
    names=["Length", "Diameter", "Height", "Whole weight", "Shucked weight",
           "Viscera weight", "Shell weight", "Age"])

abalone_train.head() # This function means "preview the first 5 lines of the loaded data"
```

```
from google.colab import drive
drive.mount('/content/gdrive')
```

```
#We reload the model that we trained in the other notebook in this folder

export_path_keras = "/content/gdrive/MyDrive/assets/abalone_model.h5" # This path may be different depending on where you stored the model (in the other notebook)
reloaded_model = tf.keras.models.load_model(
export_path_keras)
reloaded_model.summary()
```

```
# We prepare the data as we did in the original CSV Loading notebook

abalone_features = abalone_train.copy() # We avoid changing the original data by creating a copy
abalone_labels = abalone_features.pop('Age')

# We convert the features into a single numpy array
abalone_features = np.array(abalone_features)
```

```
# We now use the loaded model to train the same dataset, but with the parameters from the original model
reloaded_model.fit(abalone_features, abalone_labels, epochs=40)
```
